Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Loading Credit Card Dataset to Pandas Dataframe

```python
cc_df=pd.read_csv('creditcard.csv')
```

```python
# printing first 5 rows 
cc_df.head()
```

```python
 # printing last 5 rows
cc_df.tail()
```

```python
cc_df.info() #printing details of the dataset
```

```python
cc_df.isnull().sum() #printing missing values in each column
```

```python
# printing number of fraudulent and legit transactions
# 0- legit transaction; 1- fraudulent transaction
cc_df['Class'].value_counts()
```

```python
# Dataset is highly unbalanced 
# separating the data
legit = cc_df[cc_df.Class==0]
fraud = cc_df[cc_df.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
# printing statistical functions for column 'Amount'
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# doing comparsion between legit and fraud data
cc_df.groupby('Class').mean()
```

<!-- Now, we will be building a sample dataset containing similar distribution of both data -->

```python
legit_sample = legit.sample(n=492) #n is number of values(here, we are taking 492; equivalent to fraud data) , it will print 492 values in random order
print(legit_sample)
```

```python
# concatenating two dataframes into new dataset
new_ds = pd.concat([legit_sample, fraud], axis=0) #axis = 0 means, values will be added row-wise
```

```python
new_ds.head() #printing first 5 rows from new dataset
```

```python
new_ds.tail() #printing last 5 rows 
```

```python
new_ds['Class'].value_counts()
```

```python
new_ds.groupby('Class').mean() #comparison of new dataset
```

Splitting the data into features and targets

```python
#dropping column 'Class' and storing in y
x=new_ds.drop(columns='Class', axis=1) 
y=new_ds['Class']
```

```python
print(x)
```

```python
print(y)
```

Splitting data into Training and Testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, stratify=y, random_state=2)
#20% of data goes to testing data and rest 80% to training data
```

```python
print(x.shape, x_test.shape, x_train.shape)
```

Model Training using Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

Evaluating model using Accuracy Score

```python
# Training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print("Accuracy score on training data:",training_data_accuracy)
```

```python
# Testing data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print("Accuracy score on testing data:",test_data_accuracy)
```
